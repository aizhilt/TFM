```python
import pandas as pd
```

```python
data = pd.read_csv('parque_vehiculos_2022.csv', encoding='latin-1', sep=';')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.columns
```

```python
# Filtrar
columnas_deseadas = ["DISTRITO", "CONTADOR"]
df_filtrado = df[columnas_deseadas]
```

```python
resultados = df_filtrado.groupby("DISTRITO")["CONTADOR"].sum().reset_index()
```
